```python
import pandas as pd
import json

sheet = pd.read_csv('data/resources.csv')
sheet.head()
```

```python
root = {"name": "root", "children": []}

for _, row in sheet.iterrows():
    category = str(row["Category"]).strip()
    location = str(row["Location"]).strip()
    resource = str(row["Resource"]).strip()

    # Find or create category node
    cat_node = next((c for c in root["children"] if c["name"] == category), None)
    if not cat_node:
        cat_node = {"name": category, "children": []}
        root["children"].append(cat_node)

    # Find or create location node
    loc_node = next((l for l in cat_node["children"] if l["name"] == location), None)
    if not loc_node:
        loc_node = {"name": location, "children": []}
        cat_node["children"].append(loc_node)

    # Add resource node
    loc_node["children"].append({"name": resource, "value": 1})

# Save JSON
with open("data/resources.json", "w") as f:
    json.dump(root, f, indent=2)

print("✅ resources.json created!")
```
